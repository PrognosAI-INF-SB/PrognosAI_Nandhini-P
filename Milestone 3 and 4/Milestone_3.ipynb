```python
"""
AI-PrognosAI: Predictive Maintenance System
Complete Implementation - Milestones 1-4
Hybrid LSTM Model for Remaining Useful Life (RUL) Prediction
"""

# ============================================================================
# SETUP & IMPORTS
# ============================================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, callbacks
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("=" * 80)
print("AI-PrognosAI: Predictive Maintenance System")
print("Hybrid LSTM Model for RUL Prediction")
print("=" * 80)

# ============================================================================
# MILESTONE 1: DATA PREPARATION & FEATURE ENGINEERING
# ============================================================================
print("\n" + "=" * 80)
print("MILESTONE 1: DATA PREPARATION & FEATURE ENGINEERING")
print("=" * 80)

# Load dataset
try:
    df = pd.read_csv('sensor_data.csv')
    print("\n✓ Dataset loaded successfully from 'sensor_data.csv'")
except FileNotFoundError:
    print("\n✗ File 'sensor_data.csv' not found. Using simulated data...")
    # Simulate dataset with characteristics matching the description
    n_samples = 140000
    time_vals = np.arange(n_samples)
    df = pd.DataFrame({
        'time': time_vals,
        'SensorA': 9.0 + np.sin(time_vals/1000) * 2 + np.random.randn(n_samples) * 1.5,
        'SensorB': 4.0 + np.cos(time_vals/800) * 3 + np.random.randn(n_samples) * 2.0,
        'SensorC': 8.0 - (time_vals/n_samples) * 2 + np.random.randn(n_samples) * 1.2
    })
    # Add some missing values randomly
    mask = np.random.rand(n_samples) > 0.98
    df.loc[mask, 'SensorA'] = np.nan
    mask = np.random.rand(n_samples) > 0.98
    df.loc[mask, 'SensorB'] = np.nan

# Dataset Summary
print("\n--- Dataset Summary ---")
print(f"Shape: {df.shape}")
print(f"\nColumn Types:\n{df.dtypes}")
print(f"\nMissing Values:\n{df.isnull().sum()}")
print(f"\nDescriptive Statistics:\n{df.describe()}")

# Data Cleaning
print("\n--- Data Cleaning ---")
# Convert time column if it's in string format
if df['time'].dtype == 'object':
    df['time'] = pd.to_datetime(df['time'], format='%H:%M', errors='coerce')
    df['time'] = (df['time'] - df['time'].min()).dt.total_seconds() / 60
else:
    df['time'] = df['time'].astype(float)

# Handle missing values - forward fill then backward fill
sensor_cols = ['SensorA', 'SensorB', 'SensorC']
for col in sensor_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col].fillna(method='ffill', inplace=True)
    df[col].fillna(method='bfill', inplace=True)

print(f"✓ Missing values handled")
print(f"Remaining missing values: {df[sensor_cols].isnull().sum().sum()}")

# Normalize sensor data
print("\n--- Feature Normalization ---")
scaler = StandardScaler()
df[sensor_cols] = scaler.fit_transform(df[sensor_cols])
print(f"✓ Sensor columns normalized (StandardScaler)")
print(f"Mean after scaling: {df[sensor_cols].mean().values}")
print(f"Std after scaling: {df[sensor_cols].std().values}")

# Create rolling window sequences
print("\n--- Creating Rolling Window Sequences ---")
WINDOW_SIZE = 30  # timesteps
STEP_SIZE = 5     # stride for creating sequences

def create_sequences(data, window_size, step_size):
    """Create 3D sequences (samples, timesteps, features)"""
    X, indices = [], []
    for i in range(0, len(data) - window_size, step_size):
        X.append(data[i:i + window_size])
        indices.append(i + window_size)
    return np.array(X), np.array(indices)

# Prepare feature matrix
feature_data = df[sensor_cols].values
X_sequences, sequence_indices = create_sequences(feature_data, WINDOW_SIZE, STEP_SIZE)

print(f"✓ Sequences created")
print(f"  Input shape (samples, timesteps, features): {X_sequences.shape}")

# Compute RUL targets
print("\n--- Computing RUL Targets ---")
# RUL decreases linearly from max to 0 at the end
max_rul = 100
total_cycles = len(df)
rul_full = max_rul * (1 - df['time'].values / df['time'].max())
rul_full = np.maximum(rul_full, 0)  # Ensure non-negative

# Extract RUL for each sequence (RUL at the end of the window)
y_rul = rul_full[sequence_indices]

print(f"✓ RUL targets computed")
print(f"  RUL shape: {y_rul.shape}")
print(f"  RUL range: [{y_rul.min():.2f}, {y_rul.max():.2f}]")
print(f"  RUL mean: {y_rul.mean():.2f} ± {y_rul.std():.2f}")

# Train/Validation/Test Split
print("\n--- Data Splitting ---")
# Split: 70% train, 15% validation, 15% test
X_temp, X_test, y_temp, y_test = train_test_split(
    X_sequences, y_rul, test_size=0.15, random_state=42, shuffle=True
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.176, random_state=42, shuffle=True  # 0.176 of 85% ≈ 15%
)

print(f"✓ Data split completed")
print(f"  Training set:   {X_train.shape[0]} samples ({X_train.shape[0]/len(X_sequences)*100:.1f}%)")
print(f"  Validation set: {X_val.shape[0]} samples ({X_val.shape[0]/len(X_sequences)*100:.1f}%)")
print(f"  Test set:       {X_test.shape[0]} samples ({X_test.shape[0]/len(X_sequences)*100:.1f}%)")

print("\n" + "=" * 80)
print("MILESTONE 1 COMPLETE ✓")
print("=" * 80)

# ============================================================================
# MILESTONE 2: MODEL DEVELOPMENT (TRAINING)
# ============================================================================
print("\n" + "=" * 80)
print("MILESTONE 2: HYBRID LSTM MODEL DEVELOPMENT")
print("=" * 80)

print("\n--- Building Hybrid LSTM Architecture ---")

def build_hybrid_lstm(input_shape):
    """
    Hybrid LSTM combining:
    - 1D Convolutional layers for local pattern extraction
    - Bidirectional LSTM for temporal dependencies
    """
    model = models.Sequential([
        # Input layer
        layers.Input(shape=input_shape),

        # Convolutional feature extraction
        layers.Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Dropout(0.2),

        layers.Conv1D(filters=32, kernel_size=3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Dropout(0.2),

        # Bidirectional LSTM layers
        layers.Bidirectional(layers.LSTM(64, return_sequences=True)),
        layers.Dropout(0.3),

        layers.Bidirectional(layers.LSTM(32, return_sequences=False)),
        layers.Dropout(0.3),

        # Dense layers for regression
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(32, activation='relu'),
        layers.Dense(1, activation='linear')  # RUL output
    ])

    return model

# Build model
input_shape = (WINDOW_SIZE, len(sensor_cols))
model = build_hybrid_lstm(input_shape)

# Compile model
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae', 'mse']
)

print("✓ Hybrid LSTM model built")
print("\n--- Model Architecture ---")
model.summary()

# Setup callbacks
print("\n--- Setting up Training Callbacks ---")
early_stop = callbacks.EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6,
    verbose=1
)

checkpoint = callbacks.ModelCheckpoint(
    'best_hybrid_lstm_model.h5',
    monitor='val_loss',
    save_best_only=True,
    verbose=0
)

print("✓ Callbacks configured: EarlyStopping, ReduceLROnPlateau, ModelCheckpoint")

# Train model
print("\n--- Training Hybrid LSTM Model ---")
EPOCHS = 100
BATCH_SIZE = 64

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[early_stop, reduce_lr, checkpoint],
    verbose=1
)

print("\n✓ Training completed")

# Plot training history
print("\n--- Training History Visualization ---")
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loss plot
axes[0].plot(history.history['loss'], label='Training Loss', linewidth=2)
axes[0].plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
axes[0].set_xlabel('Epoch', fontsize=12)
axes[0].set_ylabel('Loss (MSE)', fontsize=12)
axes[0].set_title('Model Loss Over Epochs', fontsize=14, fontweight='bold')
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)

# MAE plot
axes[1].plot(history.history['mae'], label='Training MAE', linewidth=2)
axes[1].plot(history.history['val_mae'], label='Validation MAE', linewidth=2)
axes[1].set_xlabel('Epoch', fontsize=12)
axes[1].set_ylabel('Mean Absolute Error', fontsize=12)
axes[1].set_title('Model MAE Over Epochs', fontsize=14, fontweight='bold')
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "=" * 80)
print("MILESTONE 2 COMPLETE ✓")
print("=" * 80)

# ============================================================================
# MILESTONE 3: EVALUATION & PERFORMANCE ASSESSMENT
# ============================================================================
print("\n" + "=" * 80)
print("MILESTONE 3: EVALUATION & PERFORMANCE ASSESSMENT")
print("=" * 80)

# Predictions on test set
print("\n--- Generating Predictions on Test Set ---")
y_pred = model.predict(X_test, verbose=0).flatten()

# Calculate RMSE
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\n--- Performance Metrics ---")
print(f"Root Mean Square Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE):     {mae:.4f}")
print(f"R² Score:                      {r2:.4f}")

# Interpretation
print(f"\n--- Model Interpretation ---")
print(f"✓ Average prediction error: ±{rmse:.2f} cycles")
print(f"✓ Model explains {r2*100:.2f}% of variance in RUL")
if rmse < 10:
    print("✓ Excellent performance: RMSE < 10 cycles")
elif rmse < 20:
    print("✓ Good performance: RMSE < 20 cycles")
else:
    print("⚠ Moderate performance: Consider model improvements")

# Visualization: Actual vs Predicted
print("\n--- Actual vs Predicted RUL Visualization ---")
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Scatter plot
axes[0].scatter(y_test, y_pred, alpha=0.5, s=20, edgecolors='k', linewidth=0.5)
axes[0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()],
             'r--', lw=2, label='Perfect Prediction')
axes[0].set_xlabel('Actual RUL', fontsize=12)
axes[0].set_ylabel('Predicted RUL', fontsize=12)
axes[0].set_title(f'Actual vs Predicted RUL\nRMSE = {rmse:.4f}',
                  fontsize=14, fontweight='bold')
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)

# Time series comparison (first 200 samples)
sample_size = min(200, len(y_test))
axes[1].plot(y_test[:sample_size], label='Actual RUL', linewidth=2, alpha=0.7)
axes[1].plot(y_pred[:sample_size], label='Predicted RUL', linewidth=2, alpha=0.7)
axes[1].set_xlabel('Sample Index', fontsize=12)
axes[1].set_ylabel('RUL (cycles)', fontsize=12)
axes[1].set_title('RUL Prediction Comparison (First 200 Samples)',
                  fontsize=14, fontweight='bold')
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Error distribution
print("\n--- Prediction Error Distribution ---")
errors = y_pred - y_test
plt.figure(figsize=(10, 5))
plt.hist(errors, bins=50, edgecolor='black', alpha=0.7)
plt.axvline(0, color='red', linestyle='--', linewidth=2, label='Zero Error')
plt.xlabel('Prediction Error (Predicted - Actual)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title(f'Prediction Error Distribution\nMean Error: {errors.mean():.4f}, Std: {errors.std():.4f}',
          fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print("\n" + "=" * 80)
print("MILESTONE 3 COMPLETE ✓")
print("=" * 80)
```
